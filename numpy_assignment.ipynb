```python
import numpy as np
arr=np.random.rand(5,4)
print("original array:")
print(arr)
sums=arr.sum(axis=1)
print("row_sum:")
print(sums)
row_max=np.argmax(sums)
print("row_max:")
print(row_max)
print(arr[max_row])
normalized=np.empty_like(arr)
for i in range(5):
  minimum=arr[i].min()
  maximum=arr[i].max()
  normalized[i]=(arr[i]-minimum)/(minimum-maximum)
print("Normalized Array:")
print(normalized)
```

```python
arr = np.arange(1, 37).reshape(6, 6)
print("Original Array:")
print(arr)
submatrix = arr[1:5, 1:5]
print("Middle 4x4 Submatrix:")
print(submatrix)
even_numbers = submatrix[submatrix % 2 == 0]

print("Even Numbers:")
print(even_numbers)
```

```python
arr=np.arange(12).reshape(3,4)
print("Original Array:")
print(arr)
arr2=arr.reshape(2,6)
print("reshaped array:")
print(arr2)
arr3=arr2.flatten()
print("flattened array:")
print(arr3)
arr4=arr3.reshape(4,3)
print("Reshaped to (4,3):")
print(arr4)

```

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
print("original array:")
print(a)

a = np.vstack((a, [5, 6]))

print("After Adding Row:")
print(a)

a = np.hstack((a, [[7], [8], [9]]))

print("\nAfter Adding Column:")
print(a)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
arr=np.concatenate((a,b,c))
print("concatenated array:")
print(arr)
matrix=arr.reshape(3,3)
print("reshaped matrix:")
print(matrix)
diagonal=np.diag(matrix)
print("diagonal matrix:")
print(diagonal)

```

```python
arr=np.arange(1,21)
print("original array:")
print(arr)
result=arr[(arr % 3==0)|(arr % 5==0)]
print("Values divisible by 3 or 5:")
print(result)
count=len(result)
print("Count:")
print(count)
```

```python
import pandas as pd

df = pd.read_csv("/content/titanic_dataset.csv")

df['Age'] = df['Age'].fillna(df['Age'].median())

print(df['Age'])
```
